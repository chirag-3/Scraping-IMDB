```python
url = "https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page=1&ref_=adv_nxt"
```

```python
from bs4 import BeautifulSoup
import requests
```

# Printing names and genres of first 3 movies

```python
response = requests.get(url)
response
```

```python
page = response.text
soup = BeautifulSoup(page,'html.parser')

print(soup.prettify())
```

```python
movies = soup.find_all(class_='lister-item-content')
```

```python
c = 0
for movie in movies:
    print(movie.h3.a.string.strip(),end=';')
    print(movie.find(class_='genre').string.strip())
    c += 1
    if c == 3:
        break
```
